```python
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
datafile = "XPS_NMRC_CrCl2pym_Cr_2p_1"
data = np.genfromtxt(datafile,skip_header=7)
data = data[np.argsort(data[:, 0])] # sort all data by binding energy
BE = data[:,0]

fig = plt.figure(1, figsize=(3.6, 3.6))
ax1 = plt.subplot()

ax1.plot (BE,data[:,1]/1000, 'o', mfc='none', color='teal', label='data')
ax1.plot (BE,data[:,9]/1000, color='teal', label='fit')
ax1.plot (BE,data[:,2]/1000, color='red', label='Cr$^\mathrm{III}$')
ax1.plot (BE,data[:,3]/1000, color='red')
ax1.plot (BE,data[:,4]/1000, color='blue', label='Cr$^\mathrm{VI}$')
ax1.plot (BE,data[:,5]/1000, color='blue')
ax1.plot (BE,data[:,6]/1000, color='green', label='Cr$^0$')
ax1.plot (BE,data[:,7]/1000, color='green')
ax1.plot (BE,data[:,8]/1000, color='teal') #, label='background')


ax1.set_xlabel('Binding Energy (eV)',fontsize=9)
ax1.set_ylabel('CPS',fontsize=9)
ax1.set_xlim((595,566))
ax1.text(595, 3.38, '$\\times 10^3$',fontsize=7)
#ax1.set_ylim((0.5,1))
ax1.tick_params(axis='both', labelsize=8)
plt.legend(frameon=False, loc='upper right', fontsize=7)
#plt.savefig('200124_sijisi_XPS_CrpymCl2_Cr_2p_1.pdf', bbox_inches='tight', transparent=True)
plt.show()
```
