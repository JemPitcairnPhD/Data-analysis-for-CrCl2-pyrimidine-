```python
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('PXRD_CrCl2pym_riet.dat') #PXRD data file which has been converted to .dat or .xy
rug = np.genfromtxt('PXRD_CrCl2pym_rug_marks.dat')
```

```python
fig = plt.figure(1, figsize=(3.75, 2))
ax1 = plt.subplot()

Q = 4*np.pi*np.sin(data[:,0]/2/180*np.pi)/1.54

obs = ax1.scatter (Q,data[:,1]+900, s=10, label='$I_{obs}$', color='indigo')
calc, = ax1.plot (Q,data[:,2]+900, linewidth=1.5, label='$I_{calc}$', color='tomato')
diff, = ax1.plot (Q,data[:,3]+500, linewidth=1.5, label='$I_{obs}$-$I_{calc}$', color='grey')
ax1.plot(rug[:,4], ([0]*len(rug)), '|', ms=4, c='tomato')

ax1.set_xlabel('$Q$ (Ã…$^{-1}$)', fontsize=9)
ax1.set_ylabel('$I$', fontsize=9)
ax1.set_xlim((0.7,4.1))
#ax1.set_ylim((0,16000))
plt.yticks([])
plt.xticks([1,2,3,4],fontsize=8)
ax1.tick_params(axis='both', labelsize=8)
ax1.legend([obs, calc, diff],['$I_{\mathrm{obs}}$','$I_{\mathrm{calc}}$','$I_{obs}-I_{calc}$'], frameon=False, loc='upper right', fontsize=9)
plt.tight_layout()
#plt.savefig('PXRD_huxira_riet.pdf', bbox_inches='tight') #The output of your plot as an image or a pdf
plt.show()
```
